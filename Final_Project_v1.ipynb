```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (8.0, 6.0) #setting figure size
import seaborn as sns

sns.set_style('whitegrid')
```

```python
df=pd.read_csv('MFG10YearTerminationData.csv')
```

```python
df.head()
```

```python
pivot = df.pivot_table(index=['EmployeeID'], values=['length_of_service'], aggfunc='max')
```

```python
pivot.head(50)
```

```python
plt.figure()
plt.hist(pivot['length_of_service'], bins=100)
plt.xlabel('length_of_service')
plt.ylabel('EmployeeID')
plt.show()
```
