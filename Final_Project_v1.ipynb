```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (10.0, 8.0) 
import seaborn as sns

sns.set_style('whitegrid')
```

```python
df=pd.read_csv('MFG10YearTerminationData.csv')
```

```python
df.head(10)
```

#### Create a pivot table that shows the employee ID with the max status year thats avavaliable in the data set

```python
max_year = df.pivot_table(index=['EmployeeID'], values=['STATUS_YEAR'], aggfunc='max')
```

```python
max_year.head(20)
```

### did a inner join between the max status year table and df table to get the most recent data for each employee

```python
most_recent = df.merge(max_year, on=['EmployeeID','STATUS_YEAR'], how='inner')
```

```python
most_recent.head()
```

```python
plt.figure()
plt.hist(most_recent['age'], bins=26)
plt.xlabel('age')
plt.ylabel('EmployeeID')
plt.show()
```
